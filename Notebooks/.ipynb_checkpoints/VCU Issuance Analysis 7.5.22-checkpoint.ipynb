From looking at the info, the Issuance Date and retirement/cancellation date are stored as strings, this has to be cleaned and turned into DateTime objects

```python
#create a new dataset to start performing EDA

df_clean = df.copy()
```

```python
#gut check

df_clean.info()
```

```python
# Converting to Datetime

df_clean['Issuance Date'] = pd.to_datetime(df_clean['Issuance Date'])

df_clean['Vintage Start'] = pd.to_datetime(df_clean['Vintage Start'])

df_clean['Vintage End'] = pd.to_datetime(df_clean['Vintage End'])

df_clean['Retirement/Cancellation Date'] = pd.to_datetime(df_clean['Retirement/Cancellation Date'])
```

```python
df_clean[df_clean.Country == 'India']
```

```python
#how many countries are in the dataset? 78 Countries in Dataset

df_clean['Country'].value_counts()
```

```python
#create a grouped df showing the quantity of issued certs by country
df_country_issuance = df_clean.groupby(['Country']).sum()

df_country_issuance.head()
```

```python
#create a grouped df showing the quantity of issued certs by country, sorted by the largest sum to smallest
df_country_issuance = df_country_issuance.sort_values(by=['Quantity Issued'],
                                                     ascending=False)


df_country_issuance.head()
```

```python
#reset the index/gut check

df_country_issuance.reset_index(inplace=True)

df_country_issuance
```

```python
#create a grouped df showing the quantity of issued certs by type, sorted by the largest sum to smallest and reset index
df_project_issuance = df_clean.groupby(['Project Type']).sum()

df_project_issuance = df_project_issuance.sort_values(by=['Quantity Issued'],
                                                     ascending=False)
df_project_issuance.reset_index(inplace=True)

df_project_issuance.head()
```

```python
###create a grouped df into pivot table showing the quantity of issued certs by type, country
# - sorted by the largest sum to smallest and reset index
country_project = df_clean.groupby(['Country', 'Project Type']).sum()

country_project = country_project.sort_values(by=['Quantity Issued', 'Country'], ascending=False)

country_project.reset_index(inplace=True)

country_project.head()
```

```python
#attempting to make a bar chart showing issued credits by country

sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(10, 16))

sns.set_color_codes("pastel")
plot = sns.barplot(x="Quantity Issued", y="Country", data=df_country_issuance)

ax.ticklabel_format(style='sci', scilimits=(0,20), axis='x')
ax.set(xlim=(0, 210000000),
       ylabel="",
       ylim=(0,14),
       xlabel="")

xlabels = ['{:,.0f}'.format(x) + ' MM' for x in plot.get_xticks()/1000000]
plot.set_xticklabels(xlabels)

sns.despine(left=True, bottom=True)
```

```python
#repeat for project type
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(10, 16))

sns.set_color_codes("pastel")
plot = sns.barplot(x="Quantity Issued", y="Project Type", data=df_project_issuance)

ax.ticklabel_format(style='sci', scilimits=(0,20), axis='x')
ax.set(xlim=(0, 500000000),
       ylabel="",
       ylim=(0,10),
       xlabel="")

xlabels = ['{:,.0f}'.format(x) + ' MM' for x in plot.get_xticks()/1000000]
plot.set_xticklabels(xlabels)

sns.despine(left=True, bottom=True)
```

```python
df_clean['Time to Retirement'] = df_clean['Retirement/Cancellation Date'] - df_clean['Issuance Date']

df_clean.head()
```

```python
df_clean.columns
```

```python
df_issuance = df_clean.copy()

df_issuance.head()
```

```python
df_issuance= df_issuance.dropna(subset=['Time to Retirement'])

df_issuance.head()
```

```python
df_issuance['Int Time to Retirement'] = df_issuance['Time to Retirement'].astype('timedelta64[D]')

df_issuance.info()
```

```python
retirement_time = df_issuance.groupby('Project Type').mean()

retirement_time = retirement_time.sort_values(by=['Int Time to Retirement'],
                                              ascending=True)
retirement_time.reset_index(inplace=True)

retirement_time.head()
```

```python
# let's check time to retirement for credits issued 2018 onward
#create dataframe of 2018 on

retired_2018_on = df_issuance.loc[df_issuance['Retirement/Cancellation Date'].dt.year >= 2018]

retired_2018_on.head()
```

```python
retired_2018_on = retired_2018_on.groupby('Project Type').mean()

retired_2018_on = retired_2018_on.sort_values(by=['Int Time to Retirement'],
                                              ascending=True)
retired_2018_on.reset_index(inplace=True)

retired_2018_on.head()
```

```python
#what about for credits issued 2018 on?

issued_2018_on = df_issuance.loc[df_issuance['Issuance Date'].dt.year >= 2018]

issued_2018_on = issued_2018_on.groupby('Project Type').mean()

issued_2018_on = issued_2018_on.sort_values(by=['Int Time to Retirement'],
                                              ascending=True)
issued_2018_on.reset_index(inplace=True)

issued_2018_on.head()
```

```python
#Last one - credits retired 2021 on 

retired_2021_on = df_issuance.loc[df_issuance['Retirement/Cancellation Date'].dt.year >= 2021]

retired_2021_on = retired_2021_on.groupby('Project Type').mean()

retired_2021_on = retired_2021_on.sort_values(by=['Int Time to Retirement'],
                                              ascending=True)
retired_2021_on.reset_index(inplace=True)

retired_2021_on.head()
```

```python
#Last one - credits issued 2021 on 

issued_2021_on = df_issuance.loc[df_issuance['Issuance Date'].dt.year >= 2021]

issued_2021_on = issued_2021_on.groupby('Project Type').mean()

issued_2021_on = issued_2021_on.sort_values(by=['Int Time to Retirement'],
                                              ascending=True)
issued_2021_on.reset_index(inplace=True)

issued_2021_on.head()
```

```python
#Beneficiary Analysis

beneficiary = df_issuance.loc[df_issuance['Retirement Beneficiary'].notnull()]

beneficiary.head()
```

```python
import numpy as np

df_beneficiary = beneficiary.groupby('Retirement Beneficiary').agg({'Quantity Issued':'sum',
                                                                    'Int Time to Retirement':'mean'})

df_beneficiary = df_beneficiary.sort_values(by=['Quantity Issued'],
                                              ascending=False)
df_beneficiary.reset_index(inplace=True)

df_beneficiary.head(20)
```

```python
sns.set_theme(style='darkgrid',
             palette='deep')

f, ax = plt.subplots(figsize=(10, 16))

plot = sns.barplot(x="Quantity Issued", y="Retirement Beneficiary", data=df_beneficiary)

ax.ticklabel_format(style='sci', scilimits=(0,20), axis='x')
ax.set(xlim=(0, 15000000),
       ylabel="",
       ylim=(0,21),
       xlabel="")

xlabels = ['{:,.0f}'.format(x) + ' MM' for x in plot.get_xticks()/1000000]
plot.set_xticklabels(xlabels)

sns.despine(left=True, bottom=True)
```
